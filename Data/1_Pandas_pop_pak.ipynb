```python
# Rural population of pakistan in 2017 was 
print("Rural population of pakistan in 2017 was"), df['ALL SEXES (RURAL)'].sum()
```

```python
pop_rural1998 = df['POPULATION 1998 (RURAL)'].sum()
pop_rural2017 = df['ALL SEXES (RURAL)'].sum()
pop_changeRural = (pop_rural2017-pop_rural1998)/pop_rural1998*100
pop_changeRural
```

```python
print("the population in 2017 has increased this number 41632809")
```

```python
pop_urban1998 = df['POPULATION 1998 (URBAN)'].sum()
pop_urban2017 = df['ALL SEXES (URBAN)'].sum()
pop_changeUrban = (pop_urban2017-pop_urban1998)/pop_urban1998* 100
pop_changeUrban
```

# Caculate the % of population

```python
print('Percent change of Population in Rural Areas', pop_changeRural,"%  from 1998 to 2017")
print('Percent change of Population in Urban Areas', pop_changeUrban, "%  from 1998 to 2017")
```
